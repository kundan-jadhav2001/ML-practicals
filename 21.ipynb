```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the "Mall_Customers.csv" dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
# Select the relevant features
X = df[['Age', 'Spending Score (1-100)']]
```

```python
# Calculate the within-cluster sum of squares (WCSS) for different values of k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS for different values of k
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Create a K-Means clusterer with the optimal k value (e.g., k=3)
kmeans_optimal = KMeans(n_clusters=3, init='k-means++', random_state=42)

# Fit the data and predict cluster labels
cluster_labels = kmeans_optimal.fit_predict(X)
```
