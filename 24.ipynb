```python
# Use Wine data set on UCI Machine Learning Repository.
# Visualize the data in 2 Dimensional space. Apply feature
# scaling on it and again visualize the same in 2-D and comment
# on the results.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
```

```python
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
```

```python
# Select the first two features for visualization
X_2D = X[:, :2]

# Create a scatter plot to visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X_2D[:, 0], X_2D[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.title("Wine Data Visualization in 2-D (Before Scaling)")
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_2D_scaled = X_scaled[:, :2]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_2D_scaled[:, 0], X_2D_scaled[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel("Scaled " + feature_names[0])
plt.ylabel("Scaled " + feature_names[1])
plt.title("Wine Data Visualization in 2-D (After Scaling)")
plt.show()
```
